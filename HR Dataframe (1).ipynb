```python
custom_palette = ["brown","yellow"]
sns.set_style("whitegrid")
sns.countplot(x='Attrition', data=hr, palette=custom_palette)
plt.text(x=0.5, y=-210, s='Results: Data of Attrition says that it has more no of No values and less no of Yes values', fontsize=17, color='black', ha='center')
plt.show( )
```

```python
custom_palette = ["brown","black"]
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6)) 
sns.countplot(hue=hr.Attrition,x=hr.BusinessTravel, palette=custom_palette)
plt.text(x=0.5, y=-160, s='Results: Data of BusinessTravel shows that company has more no of employes who travels rarely', fontsize=18, color='black', ha='center')
plt.show( )
```

```python
custom_palette = ["red","white"]
sns.set_style("dark")
plt.figure(figsize=(8, 6)) 
sns.countplot(hue=hr.Attrition,x=hr.Department, palette=custom_palette)
plt.text(x=0.5, y=-160, s='Results: Research and Developement department have more no of Attrition as compared to other two departments', fontsize=19, color='blue', ha='center')
plt.show( )
```
